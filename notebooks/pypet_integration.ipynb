```python
from pypet import Environment
env = Environment(trajectory='Diffusion-0-2',
                  comment='Experiment to simulate diffusion on '
                        'hypergraphs. ',
                  add_time=False, # We don't want to add the current time to the name,
                  log_folder='./logs/',
                  multiproc=True,
                  ncores=2, # My laptop has 2 cores ;-)
                  filename='./hdf5/', # We only pass a folder here, so the name is chosen
                  # automatically to be the same as the Trajectory
                  )
```

```python
traj = env.v_trajectory
```

```python
traj.f_add_parameter('hypergraph.number_of_nodes', 10,
                     comment='Number of nodes in hypergraph')
traj.f_add_parameter('hypergraph.number_of_hyperedges', 4,
                     comment='Number of hyperedges in hypergraph')
traj.f_add_parameter('hypergraph.cardinality', 4,
                     comment='Number of nodes in a hyperedge')
traj.f_add_parameter('simulation.duration', 1000.0,
                     comment='The duration of the experiment in '
                            'milliseconds.')
```

```python
from pypet.utils.explore import cartesian_product
import numpy as np

explore_dict = {'hypergraph.number_of_nodes': np.arange(4, 20,1).tolist(),
                'hypergraph.number_of_hyperedges': [5, 7, 10, 20]}

explore_dict = cartesian_product(explore_dict, ('hypergraph.number_of_nodes', 'hypergraph.number_of_hyperedges'))
# The second argument, the tuple, specifies the order of the cartesian product,
# The variable on the right most side changes fastest and defines the
# 'inner for-loop' of the cartesian product

traj.f_explore(explore_dict)
```

```python
from numpy import linalg as LA

from hypergraph.markov_diffusion import create_markov_matrix_model_nodes
from hypergraph import generators
from hypergraph.diffusion_engine import DiffusionEngine
from hypergraph import utils
from functools import partial
from hypergraph.analytical import prediction


#TODO
# all imports [done]
# storing data in traj [done]
# some comments, what is happening there
# analysis of what was just generated

def run_diffusion_node(traj):
    print('oh!')
    HG = generators.uniform_hypergraph(
        traj.par.hypergraph.number_of_nodes,
        traj.par.hypergraph.number_of_hyperedges,
        traj.par.hypergraph.cardinality)
    
    t_max = 100000
    markov_matrix = create_markov_matrix_model_nodes(HG)
    engine = DiffusionEngine(markov_matrix, t_per_walker=int(t_max))
    freqs_matrix = LA.matrix_power(markov_matrix, 20)[0]
    frequencies, states = engine.simulate(t_max)
    connected_elements = [freq[0] + 1 for freq in frequencies]
    
    
    missing_nodes = set(HG.nodes()) - set(connected_elements)
    for missing_node in missing_nodes:
        frequencies.append((missing_node, 0))
    
    frequencies = [(node, frequency) for node, frequency in frequencies]
    frequencies.sort(key=lambda x: x[0])
    xs, ys = zip(*frequencies)
    
    ys = np.array(ys, dtype='float')
    ys /= sum(ys)
    width = 0.2  

    ys_prediction = partial(prediction, model='hypergraph')(HG)
    traj.f_add_result('hypergraph.01', ys=ys, xs=xs,
                      comment='Contains results of simulating diffusion')
    return (traj.par.hypergraph.number_of_nodes,
        traj.par.hypergraph.number_of_hyperedges,
        traj.par.hypergraph.cardinality, xs, ys)
    
```

```python
env.f_run(run_diffusion_node)
```

```python

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

from pypet import Environment, cartesian_product
from pypet import pypetconstants


def multiply(traj):
    """Sophisticated simulation of multiplication"""
    z=traj.x*traj.y
    traj.f_add_result('z',z, comment='I am the product of two reals!')


# Create an environment that handles running
env = Environment(trajectory='Example10',filename='experiments/example_08/HDF5/example_10.hdf5',
                  file_title='Example10', log_folder='experiments/example_08/LOGS/',
                  comment='Another example!')

# Get the trajectory from the environment
traj = env.v_trajectory

# Add both parameters
traj.f_add_parameter('x', 1, comment='I am the first dimension!')
traj.f_add_parameter('y', 1, comment='I am the second dimension!')

# Explore the parameters with a cartesian product:
x_length = 15
y_length = 15
traj.f_explore(cartesian_product({'x':range(x_length), 'y':range(y_length)}))

# Run the simulation
env.f_run(multiply)

# We load all results
traj.f_load(load_results=pypetconstants.LOAD_DATA)

# We access the ranges for plotting
xs = traj.f_get('x').f_get_range()
ys = traj.f_get('y').f_get_range()

# Now we want to directly get all numbers z from all runs
# for plotting.
# We use `fast_access=True` to directly get access to
# the values.
# Moreover, since `f_get_from_runs` returns an ordered dictionary
# `values()` gives us all values already in the correct order of the runs.
zs = traj.f_get_from_runs(name='z',fast_access=True).values()

# Convert the lists to numpy 2D arrays
x_mesh = np.reshape(np.array(xs),(x_length, y_length))
y_mesh = np.reshape(np.array(ys),(x_length, y_length))
z_mesh = np.reshape(np.array(zs),(x_length, y_length))

# Make fancy 3D plot
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x_mesh, y_mesh, z_mesh, rstride=1, cstride=1)
plt.show()
```
